```julia
using CSV, DataFrames, BioSequences, FASTX, Plots, Plots.Measures, Serialization, Distributions
```

```julia
unedited_cp = deserialize("unedited_cp.bin")
start_triplet_cp = deserialize("start_triplet_cp.bin")

labels = ["unedited_cp C→T","Non-ACG C→T","ACG C→T"]

resolution = 20
ancestral = range(0.35,0.45; length = resolution)
searchspace = logrange(1e-4, 1e-2, resolution)

plot(;legendposition = :bottomleft, tickfontsize = 8, size = (400, 400),
    xguide = "transition rate (events⋅site⁻¹⋅Ma⁻¹)", fontfamily = "Helvetica", guidefontfamily = "Helvetica Bold", guidefontsize = 9, grid = false, bottom_margin = 10mm)
plot!(; xscale = :log10, xlims = (1e-4, 1e-2), yguide = "log-likelihood", left_margin = 10mm)

plot!(searchspace, maximum.(eachslice(unedited_cp, dims = 2)); lw = 3, alpha = 0.5, label = labels[1])
plot!(searchspace, maximum.(eachslice(start_triplet_cp, dims = 2)); lw = 3, alpha = 0.5, label = labels[2])
plot!(searchspace, maximum.(eachslice(start_triplet_cp, dims = 3)); lw = 3, alpha = 0.5, label = labels[3])

plot!()
```

```julia
LL_unpartitioned = maximum(unedited_cp)
LL_partitioned = maximum(start_triplet_cp)
LRT_statistic =  2*(LL_partitioned - LL_unpartitioned)
pvalue = ccdf(Chisq(1), LRT_statistic)
println(LL_unpartitioned, "\t", LL_partitioned, "\t", pvalue)
```
